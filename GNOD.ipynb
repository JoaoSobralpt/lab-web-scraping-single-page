```python
#storing url
popvortex = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# 3. download html with a get request
response = requests.get(popvortex)
response.status_code
```

```python
#parsing
soup = BeautifulSoup(response.content, "html.parser")
```

```python
#soup
```

```python
# Find all the <p> elements with class "title-artist"
title_artist_elements = soup.find_all("p", {"class": "title-artist"})

# Initialize lists to store song titles and artists separately
song_titles = []
song_artists = []

# Iterate through each title-artist element
for element in title_artist_elements:
    # Extract the song title and artist from the element
    song_title = element.find("cite", {"class": "title"}).text.strip()
    artist = element.find("em", {"class": "artist"}).text.strip()

    # Append the extracted data to their respective lists
    song_titles.append(song_title)
    song_artists.append(artist)

# Print the titles and artists separately
for idx, (title, artist) in enumerate(zip(song_titles, song_artists)):
    print(f" Title: {title}, Artist: {artist}")



```

```python
import pandas as pd



# Combine song titles and artists into a single list
songs = [f"Title: {title}, Artist: {artist}" for title, artist in zip(song_titles, song_artists)]

# Create a DataFrame from the combined list
song_df = pd.DataFrame({'Song': songs})

# Print the DataFrame
print(song_df)
```

```python
# Save the DataFrame to a CSV file
song_df.to_csv('song_data1.csv', index=False)
```

```python
import pandas as pd
import random

# Load the DataFrame from the CSV file
song_df = pd.read_csv('song_data1.csv')  # Use the correct CSV file name

def recommend_song():
    # Ask the user for the song artist
    input_song_artist = input("Enter the song artist: ")
    
    # Check if the input artist exists in the DataFrame
    matching_songs = song_df[song_df['Song'].str.contains(input_song_artist, case=False)]  # Use the correct column name
    
    if not matching_songs.empty:
        # Select a random song from the filtered DataFrame
        recommended_song = matching_songs.sample(n=1)
        
        # Get the song information (which contains both title and artist)
        recommended_song_info = recommended_song['Song'].values[0]  # Use the correct column name
        
        return f"We recommend: {recommended_song_info}"
    else:
        return f"We have no recommendation at this time for songs by {input_song_artist}."

# Example usage:
recommendation = recommend_song()
print(recommendation)
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Initialize default values for client_id and client_secret
client_id = ''
client_secret = ''

# Read Spotify API credentials from secrets.txt
with open('secrets.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        parts = line.strip().split(':')
        if len(parts) == 2:
            key, value = parts
            if key == 'clientid':
                client_id = value.strip()
            elif key == 'clientsecret':
                client_secret = value.strip()

# Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))
```

# Retrieving audio features

```python
# We will use song_id to retrieve the audio features of a song
chunks = [(i, i + 100) for i in range(0, len(final_df), 100)]
audio_features_list = []

for chunk in chunks:
    id_list100 = final_df['track_id'][chunk[0]:chunk[1]]
    id_list100 = [track_id for track_id in id_list100 if track_id is not None]  # Filter out NoneType values
    
    if id_list100:  # Check if the list is not empty
        chunk_audio_features = sp.audio_features(id_list100)
        chunk_audio_features = [af for af in chunk_audio_features if af is not None]  # Filter out NoneType audio features
        audio_features_list += chunk_audio_features
    
    sleep(randint(1, 3))  # Add a sleep to avoid overloading the Spotify API

# Convert the list of audio features dictionaries to a DataFrame
if audio_features_list:
    audio_features_df = pd.DataFrame(audio_features_list)
else:
    audio_features_df = pd.DataFrame()  # Create an empty DataFrame if there are no audio features
```

```python
# Merge the track DataFrame with audio features DataFrame
df_w_audio_ft = pd.merge(left=final_df,
                        right=audio_features_df,
                        how='inner',
                        left_on='track_id',
                        right_on='id')

# Drop duplicate columns (e.g., 'id' from the audio features DataFrame)
df_w_audio_ft.drop(columns='id', inplace=True)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import random

# Load DataFrame with audio features
df_w_audio_ft = pd.read_pickle('songs_wt_aft.pkl')

# Choose the number of clusters
num_clusters = 5

# Select the audio features for clustering
X = df_w_audio_ft[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]

# Fit K-means clustering to the data
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
df_w_audio_ft['cluster_label'] = kmeans.fit_predict(X)

# Function to recommend songs
def recommend_song(track_name, num_recommendations=5):
    # Check whether or not the input song is in the PopVortex Hot 100
    if track_name.lower() in song_df['Song'].str.lower().values:
        print(f"{track_name} is in the PopVortex Hot 100.")
        
        # Recommend another Hot 100 song
        hot_100_songs = [s for s in song_df['Song'] if s.lower() != track_name.lower()]
        random_hot_100_song = random.choice(hot_100_songs)
        print(f"Recommendation from Hot 100: {random_hot_100_song}")
    else:
        print(f"{track_name} is not in the PopVortex Hot 100.")
        
        # Find the song in the DataFrame
        song = df_w_audio_ft[df_w_audio_ft['track_name'].str.lower() == track_name.lower()]

        if song.empty:
            return "Song not found in the DataFrame."

        # Get the cluster label of the song
        cluster_label = song.iloc[0]['cluster_label']

        # Filter songs in the same cluster, excluding the input song
        cluster_songs = df_w_audio_ft[df_w_audio_ft['cluster_label'] == cluster_label]
        cluster_songs = cluster_songs[cluster_songs['track_name'].str.lower() != track_name.lower()]

        # Randomly select recommendations from the cluster
        recommendations = cluster_songs.sample(num_recommendations)

        print("You like:", track_name)
        print("Here are some song recommendations:")
        
        for _, row in recommendations.iterrows():
            recommended_song = row['track_name'].title()
            artist = ', '.join(row['artists']).title()
            print(f"{recommended_song} by {artist}")

# Input the song name for recommendations
track_name = input('Please enter a track name for recommendations: ')

# Call the recommend_song function
recommend_song(track_name)
```
